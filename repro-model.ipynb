```python
# Figure 3: Vertical distribution of relative humidity
# (Mastenbrook, 1963; Murgatroyd, 1960; Telegadas and London, 1954)

relative_humidity_fig3 = np.arange(11)*0.077
pressure_fig3 = np.arange(11) * 100
plt.plot(relative_humidity_fig3, pressure_fig3)
plt.xlabel("RELATIVE HUMIDITY")
plt.xlim(0,1)
plt.ylabel("PRESSURE (MB)")
plt.ylim(1000, 0)
```

```python
# Figure 4: Vertical distribution of ozone at 35N 
# Ozone amount: 10^4 cm/km at NTP
# Altitude: height in km
height_pts_fig4 = np.array([ 0,  5, 10, 15, 20, 22.5, 25, 30, 35, 40, 45, 50])
ozone_pts_fig4 = np.array([25, 25, 40, 75, 180, 190, 160, 110, 50, 25, 10, 0])
plt.plot(ozone_pts_fig4, height_pts_fig4)
plt.xlabel("OZONE AMOUNT (10^{4}cm/km, at N.T.P.)")
plt.ylabel("HEIGHT (km)")
plt.xlim(0,300)
plt.ylim(0,50)
plt.legend(["APR. (35N)"])
```
